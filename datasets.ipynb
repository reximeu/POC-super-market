```python
import glob
import pandas
import seaborn as sns
import matplotlib.pyplot as plt, mpld3
import plotly.graph_objects as go
```

```python
path_datasets_dir = r'./datasets'
path_datasets_files = glob.glob(path_datasets_dir + "/*.csv")
path_datasets_files
```

```python
dfs = []

for path_datasets_file in path_datasets_files:
    df_temp = pandas.read_csv(path_datasets_file)
    dfs.append(df_temp)

df = pandas.concat(dfs)
df = df.sort_values('current_date')
df
```

```python
path_graphs_dir = r'./graphs'

categories_ids = df.category_id.unique()

for category_id in categories_ids:
    df_category = df[df.category_id == category_id]
    category_name = df_category.category_name.iloc[0]
    products_ids = df_category.product_id.unique()

    fig = go.Figure()

    for product_id in products_ids:
        df_product = df[df.product_id == product_id]
        product_name = df_product.product_name.iloc[0]

        x = df_product.current_date
        y = df_product.price_unit
        fig.add_trace( go.Scatter( x=x, y=y, mode='lines+markers', name= f"{product_id}_{product_name}" ) )

    fig.write_html(f"{path_graphs_dir}/{category_name}.html")
```
