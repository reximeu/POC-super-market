```python
#import the useful libraries.
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data = pd.read_csv("datasets/2022-06-03.csv")

# Printing the data head
data.head()
```

# UNIVARIATE ANALYSIS


## Categorical Ordered Univariate Analysis

```python
data['category_name'].value_counts(normalize=True).plot.barh()
plt.show()
```

```python
data['subcategory_name'].value_counts(normalize=True).plot.barh()
plt.show()
```

# BIVARIATE ANALYSIS

## Numeric-Numeric Analysis

### Scatter Plot

```python
plt.scatter(data['price_iva'], data['price_unit'])
plt.show()
```

### Pair Plot

```python
sns.pairplot(data = data, vars=['price_unit', 'price_bulk', 'price_iva'])
plt.show()
```

### Correlation Matrix

```python
corr = data[['price_unit', 'price_bulk', 'price_iva']].corr()

sns.heatmap(corr, annot=True, cmap = 'Reds')
plt.show()
```
